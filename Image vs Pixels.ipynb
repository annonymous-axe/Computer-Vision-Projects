# Create a chess board

```python
rows = []
for row in range(7):
    cols = []
    if row%2 == 0:
        start_point = 1
    else:
        start_point = 0    
    for col in range(7):
        if start_point == 1:
            cols.append(start_point)
            start_point = 0
        else:
            cols.append(start_point)
            start_point = 1
    rows.append(cols)                
```

Although, imshow() showing dark & light colors, original image which is save using imwrite() is black.

As image pixels are from 0(black) to 255(white)

So, let's create a chess board balck & white.

```python
rows = []
for row in range(7):
    cols = []
    if row%2 == 0:
        start_point = 255
    else:
        start_point = 0    
    for col in range(7):
        if start_point == 255:
            cols.append(start_point)
            start_point = 0
        else:
            cols.append(start_point)
            start_point = 255
    rows.append(cols) 
rows = np.array(rows)
```

```python
cv.imwrite('chess_board.png', rows)
```

```python
#let's create a big image of black shades
dim = (255, 1200)
ls = []
for i in range(dim[0]):
    temp = []
    for j in range(dim[1]):
        temp.append(i*1)
    ls.append(temp)
img_arr = np.array(ls)
```

```python
cv.imwrite('black_shades_2.png', img_arr)
```

# RGB Image

```python
image = cv.imread('rgb.png')
```

```python
plt.imshow(image)
```

This is different image then original, because cv convert image (in imread() as well as imwrite()) RGB to BGR or vice-versa

```python
# Let's convert to RGB from BGR
conv_img = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
# Original image
plt.imshow(conv_img)
```

```python
conv_img.shape
```

## Extract frame

```python
r = conv_img[:, :, 0]
```

```python
plt.imshow(r)
```
