```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Apply thresholding

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    gray_thr_1 = gray.copy()
    
    gray_thr_1[0gray_thr_1 > 127] = 255 # White background
    
    gray_thr_2 = gray.copy()
    gray_thr_2[gray_thr_2 > 127] = 20 # Black background
    
    cv.imshow("Original Image", img)
    cv.imshow("Gray Image"    , gray)
    cv.imshow("Gray Image Threshold-1"    , gray_thr_1)
    
    key = cv.waitKey(1)
    
    if key == 13:
        cam.release()
        break
```

# Masking 

```python
cam = cv.VideoCapture(0)

lower = np.array([190, 190, 190])
upper = np.array([255, 255, 255])
while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    mask = cv.blur(img, [2,2])
    mask = cv.inRange(mask, lower, upper)
    
    
    cv.imshow("Original Image", img)
    cv.imshow("Masking", mask)
    
    key = cv.waitKey(1)
    
    if key == 13:
        cam.release()
        break
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
mask = cv.cvtColor(mask, cv.COLOR_BGR2RGB)


plt.figure(figsize=(12,12))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("Off")

plt.subplot(1,2,2)
plt.imshow(mask)
plt.title("Masking Image")
plt.axis("Off")
```

# Masking Images

```python
def masking(img_path, lower, upper):
    
    img = cv.imread(img_path)

    blur = cv.blur(img, (4,4))
    mask = cv.inRange(blur, lower, upper)

    cv.imshow("Mask Image", mask)
    cv.waitKey(0)
```

```python
lower = np.array([100 , 60 , 30])
upper = np.array([220, 220, 220])
img_path = "elon1.jpeg"

masking(img_path, lower ,upper)
```

```python
lower = np.array([240, 240, 240])
upper = np.array([255, 255, 255])
img_path = "sandeep1.jpeg"
masking(img_path, lower, upper)
```
