```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Lemt's create a function which show the result of code

```python
def show_result(img):
    plt.imshow(img)
    plt.axis("Off")
    plt.title("Result")
```

## Load haarcascade classifier

```python
classifier = cv.CascadeClassifier("haarcascade_frontalface_default.xml")
```

# Let's detect face on Live webcame video

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    # 1.1 => Number of faces to detect
    # 8   => mNeighbors
    faces = classifier.detectMultiScale(img, 1.1, 8)
    
    try:
        for (x,y,w,h) in faces:
            cv.rectangle(img, (x,y), ((w+x), (y+h)), (0,180,0), 3)        
    
        cv.imshow("Frame", img)
    except:
        pass
    
    key = cv.waitKey(2)
    
    if key == 13:
        cam.release()
        break
```

```python
show_result(img)
```

# Display only Face window

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    # 1.1 => Number of faces to detect
    # 8   => mNeighbors
    faces = classifier.detectMultiScale(img, 1.1, 8)
    
    face_size = []
    
    try:
        for (x,y,w,h) in faces:

            cv.rectangle(img, (x,y), ((w+x), (y+h)), (0,180,0), 3)

        for face in faces:
            if face[-1] == max(faces[:, -1]):
                face_size = face
                
        x, y, w, h = face_size

        face = img[x:x+w, y:y+h]
        
    
        cv.imshow("Frame", img)
        cv.imshow("Face Frame", face)
    except:
        pass
    
    key = cv.waitKey(2)
    
    if key == 13:
        cam.release()
        break
```

```python
show_result(face)
```

# Blur face

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    # 1.1 => Number of faces to detect
    # 8   => mNeighbors
    faces = classifier.detectMultiScale(img, 1.1, 8)
    
    try:
        for (x,y,w,h) in faces:

#             cv.rectangle(img, (x,y), ((w+x), (y+h)), (0,180,0), 5)        

            img[x:x+w, y:y+h] = cv.blur(img[x:x+w, y:y+h], (32, 32))
            
            
        
        cv.imshow("Frame", img)
    except:
        pass
    
    key = cv.waitKey(2)
    
    if key == 13:
        cam.release()
        break
```

```python
show_result(img)
```

# Draw black circle on Face to hide it

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img   = cv.flip(img, 1)
    
    faces = classifier.detectMultiScale(img, 1.1, 10)
    
    try:
        for (x,y,w,h) in faces:
#             cv.rectangle(img, (x-20,y-20), (x+w+20, y+w+20), (0,180,0), 5)
            cv.circle(img, ((x+int(w/2)), (y+int(h/2))), (w+h)//2, (0, 0, 0), -2)
    except:
        pass
    
    cv.imshow("Frame", img)
    
    key = cv.waitKey(2)
    
    if key == 13:
        cv.destroyAllWindows()
        cam.release()
        break
```

```python
show_result(img)
```
