```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Acess webcam

```python
cam = cv.VideoCapture(0)

_, img = cam.read()

cv.imshow("Frame", img)

key = cv.waitKey(0)
print(key)
cam.release()
```

```python
plt.imshow(img)
```

# Working with webcam

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    cv.imshow('Frame', img)
    
    key = cv.waitKey(2)
    
    if (key == 13):
        cam.release()
        break
```

```python
plt.imshow(img)
```

## Flip image

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    cv.imshow('Frame', img)
    
    key = cv.waitKey(1)
    
    if (key == 13):
        cam.release()
        break
```

```python
plt.imshow(img)
```

# Draw Rectangle

```python
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    img = cv.rectangle(img, (200,200), (400, 400), (0, 150, 0), 5)
    
    cv.imshow('Frame', img)
    
    key = cv.waitKey(2)
    
    if(key==13):
        cam.release()
        break
```

```python
plt.imshow(img)
```

# Draw Circle

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.circle(img, (325, 238), 300, (147, 150, 237), 210)
    
    cv.imshow("Frame", img)
    
    key = cv.waitKey(2)
    
    if(key==13):
        cam.release()
        break    
```

```python
plt.imshow(img)
```

# Write text

```python
cam = cv.VideoCapture(0)

while True:
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    img = cv.putText(img, "Kamlesh", (400, 200), cv.FONT_HERSHEY_PLAIN, 2, (50, 50, 250), 2)
    
    cv.imshow("Frame", img)
    
    key = cv.waitKey(1)
    
    if key == 13:
        cam.release()
        break
```

```python
plt.imshow(img)
```
