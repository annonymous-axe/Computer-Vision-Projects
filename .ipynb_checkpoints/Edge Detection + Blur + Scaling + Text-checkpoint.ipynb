```python
import cv2 as cv
import matplotlib.pyplot as plt
```

# 1. Edge Detection

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    edge_1 = cv.Canny(img, 50, 100)
    edge_2 = cv.Canny(img, 100, 180)
    cv.imshow("Org Frame", img)
    cv.imshow("Frame 1", edge_1)
    cv.imshow("Frame 2", edge_2)
    
    if cv.waitKey(10) == 13:
        cam.release()
        break
```

```python
plt.subplot(1,3,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(edge_1)
plt.axis("off")
plt.title("Detect Edges_1")

plt.subplot(1,3,3)
plt.imshow(edge_2)
plt.axis("off")
plt.title("Detect Edges_2")
```

# 2. Blur image

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    blur_1 = cv.blur(img, (2,2))
    blur_2 = cv.blur(img, (4,4))
    blur_3 = cv.blur(img, (8,8))
    blur_4 = cv.blur(img, (16,16))
    blur_5 = cv.blur(img, (32,32))
    blur_6 = cv.blur(img, (64,64))
    blur_7 = cv.blur(img, (128,128))
    
    cv.imshow("Org Frame", img)
    cv.imshow("Blur-1", blur_1)
    cv.imshow("Blur-2", blur_2)
    cv.imshow("Blur-3", blur_3)
    cv.imshow("Blur-4", blur_4)
    cv.imshow("Blur-5", blur_5)
    cv.imshow("Blur-6", blur_6)
    cv.imshow("Blur-6", blur_7)
    
    if cv.waitKey(10) == 13:
        cam.release()
        break
```

```python
imgs = [img, blur_1, blur_2, blur_3, blur_4, blur_5, blur_6, blur_7]
size = len(imgs)

plt.figure(figsize=(12,12))
for i in range(size):
    plt.subplot(4,2,i+1)
    plt.imshow(imgs[i])
    plt.axis("off")
print("Increase bluring")
```

# Blur + Edge Detection

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    edge = cv.Canny(img, 100, 200)
    blur = cv.blur(img, (3, 3))
    blur_edge = cv.Canny(blur, 100, 200)
    
    cv.imshow("Org Frame", img)
    cv.imshow("Edge", edge)
    cv.imshow("Blur", blur)
    cv.imshow("Blur + Edge", blur_edge)
    
    if cv.waitKey(10) == 13:
        cam.release()
        break
```

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    resize_1 = cv.resize(img, (400, 400)) # Resize it 400x400
    # Resizing with respect to aspect ratio
    resize_2 = cv.resize(img, (int(img.shape[0]/2), int(img.shape[1]/2)))
    resize_3 = cv.resize(img, (int(img.shape[0]/4), int(img.shape[1]/4)))
    
    cv.imshow("Org Frame", img)
    cv.imshow("Resizing Frame-1", resize_1)
    cv.imshow("Resizing Frame-2", resize_2)
    cv.imshow("Resizing Frame-3", resize_3)
    
    if cv.waitKey(10) == 13:
        cam.release()
        break
```
