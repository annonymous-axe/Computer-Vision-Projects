# Import all neccessary libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Load and see image using matplotlib

```python
watermark_image_path = "watermark_image.jpeg"

watermark_image = cv.imread(watermark_image_path)

plt.imshow(watermark_image)
plt.axis("off")
plt.title("CV2 BGR Image")
```

<b>Convert image into RGB from BGR format</b>

```python
watermark_image = cv.cvtColor(watermark_image, cv.COLOR_BGR2RGB)

plt.imshow(watermark_image)
plt.axis("off")
plt.title("Original RGB Image")
```

<b>Masking image and find lower and upper array values mannually</b>

```python
# After many trials, I find best values as following
lower = np.array([190, 180, 200])
upper = np.array([255, 255, 255])

mask = cv.inRange(watermark_image, lower, upper)
plt.imshow(mask)
plt.axis("off")
```

# Remove watermark

<b>First use INPAINT_TELEA method of cv.inpaint()</b>

```python
dst = cv.inpaint(watermark_image,mask, 6,cv.INPAINT_TELEA)
plt.imshow(dst)
plt.axis("Off")
plt.title("Remove Background")
```

<b>First use INPAINT_NS method of cv.inpaint()</b>

```python
dst = cv.inpaint(watermark_image,mask, 3,cv.INPAINT_NS)
plt.imshow(dst)
plt.axis("Off")
plt.title("Remove Background")
```

<b>As we successfully remove watermark "Copyright".</b>
