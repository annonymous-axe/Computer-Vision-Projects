```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Brightness

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    img_1 = img * 1.8
     
    img_1[img_1 < 0  ] = 0
    img_1[img_1 > 255] = 255
    
    img_1 = img_1.astype(np.uint8)
    
    cv.imshow("Original", img)
    cv.imshow("Adjust Brightness"   , img_1)
    
    if cv.waitKey(1) == 13:
        cam.release()
        break
```

# Cool Tone

### 1. Add background color first

```python
blue = [243, 176, 37]

blue_background = []
for i in range(480):
    temp = []
    for j in range(640):
        temp.append(blue)
    blue_background.append(temp)
blue_background = np.array(blue_background).astype(np.uint8)
```

### 2. Merge background and video and add weights

```python
cam = cv.VideoCapture(0)

while True:
    
    _, img = cam.read()
    
    img = cv.flip(img, 1)
    
    
    
    merged = cv.addWeighted(img, .8, blue_background, .2, 0)
    
    cv.imshow("Original", img)
    cv.imshow("Adjust Brightness"   , merged)
    
    if cv.waitKey(1) == 13:
        cam.release()
        break
```

# Merge background <small color='red'>(Specific Color Tone)</small> in images

```python
def add_tone(img_path, color, a, b):

    img = cv.imread(img_path)
    
    background = []
    for i in range(img.shape[0]):
        temp = []
        for j in range(img.shape[1]):
            temp.append(color)
        background.append(temp)
    background = np.array(background).astype(np.uint8)

    final = cv.addWeighted(img, a, background, b, 0)

    cv.imshow("Original", img)
    cv.waitKey(0)

    cv.imshow("Proccessed", final)
    cv.waitKey(0)
```

```python
img_path = "elon.jpeg"
color = [255, 236, 155]
```

```python
add_tone(img_path, color, .8, .2)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis("Off")
plt.title("Original")


plt.subplot(1,2,2)
plt.imshow(final)
plt.axis("Off")
plt.title("Processed")
```

# Merge image with another image(/ image background)

```python
def merge(foreground_img_path, background_img_path, a, b):
    
    foreground_img = cv.imread(foreground_img_path)
    background_img = cv.imread(background_img_path)
    
    background_img = cv.resize(background_img, (foreground_img.shape[1], foreground_img.shape[0]))
    
    final_img = cv.addWeighted(foreground_img, a, background_img, b, 0)
    
    return final_img
```

```python
img = merge("elon.jpeg", "download.jpeg", .8, .4)
```

```python
cv.imshow("Image", img)
cv.waitKey(0)
```
